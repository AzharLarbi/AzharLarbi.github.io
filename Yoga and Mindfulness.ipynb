```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/Users/azhardaho/Desktop/Yoga and Meditation - Form Responses 1.csv')
df = pd.DataFrame(data)
```

```python
def yoga_age():
    #Is there a correlation between age and yoga practice?
    summary = df['Age'].describe()
    print(summary)
    
    #Histograms of ages by yoga practice
    groups = df['Are you practicing yoga/meditation everyday?'].unique() 
    plt.figure()

    #plot histogram for each group
    for group in groups:
        subset = df[df['Are you practicing yoga/meditation everyday?'] == group]['Age']
        plt.hist(subset, alpha=0.5, bins=10, label=group)

        plt.title('Yoga Practice per Age')
        plt.xlabel('Age')
        plt.ylabel('Frequency')
        plt.legend(title = 'Yoga')
        plt.show()
yoga_age()
```

Explanation-Demographics: 
    There is no clear correlation between age and practicing yoga. 
    The average age of the particpants is 20 years old. 
    Most particpants ages 18 to 21 practice Yoga sometimes. 

```python
def yoga_stress():
    #Is there a correlation between frequency of yoga practice and perceived stress levels due to work?
    #Filtering Data
    df_yoga_yes = df[df['Are you practicing yoga/meditation everyday?'] == 'Yes']

    #Plot of stress levels of those who pracctice yoga
    plt.figure()
    df_yoga_yes['Do you feel stressed due to work load?'].hist(bins=10, alpha=0.7)

    plt.title('Work Load Stress for Those who Practice Yoga')
    plt.xlabel('Workload Stress Level')
    plt.ylabel('Frequency')

    plt.show()

    #Does practicing Yoga helps medigate stress level?
    filtered_df = df[df['What do you think? will Meditation help you in reducing the stress?'].isin(['Yes', 'No', 'Maybe'])]
    response_counts = filtered_df['What do you think? will Meditation help you in reducing the stress?'].value_counts()

    #plot a bar chart of the responses
    plt.figure()
    response_counts.plot(kind='bar', color='purple', alpha=0.7)
    plt.title('Practicing Yoga and Medigating sress level')
    plt.xlabel('Response')
    plt.ylabel('Count')
    plt.show()
yoga_stress()
```

Explanation-Stress management through Yoga: 
    Of those who practice yoga, the majority of them experience work stress 
    The majority of those participants believe that Yoga helps them manage their stress level. 

Key findings: 
    Those who experience some level of stress due to work are more inclined to practice Yoga 
    This indicates that Yoga has a benefit in temrs of metigating stress levels. 
    Future research can look into the different types of Yoga and their different benefits. 
    Limitations: It would be useful to take into account the type of work participants do and to make this more
    scientifically compelling, measuring stress levels before and after doing Yoga would give great insight into it's benefits.
